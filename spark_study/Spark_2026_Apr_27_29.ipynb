CSV:
- Row-based format, no schema stored.
- Large file size, slow to read and parse.
- Worst performance for analytics.

JSON:
- Semi-structured, repeats keys → heavy.
- Slow parsing, bigger than CSV.
- Not suitable for large-scale analytics.

Parquet:
- Columnar format with compression.
- Very fast reads (predicate pushdown).
- Best for Spark analytics and ML workloads.

ORC:
- Columnar with strong compression.
- Has indexes & statistics for faster filtering.
- Great for Hive + Spark SQL environments.

Fastest Format for Further Processing:
1. Parquet (best overall for Spark)
2. ORC (best for Hive-heavy pipelines)
3. JSON (slow)
4. CSV (slowest)

```python
from pyspark.sql.functions import explode, col

df = spark.read.json("/Volumes/workspace/default/data_analysis/nested_emp_json.json")
df1 = df.withColumn("employee_exploded", explode(col("employee")))
df2 = df1.drop("employee")

# Dynamically select all fields from the exploded struct
employee_fields = df1.schema["employee_exploded"].dataType.fields
for field in employee_fields:
    df2 = df2.withColumn(field.name, col(f"employee_exploded.{field.name}"))

display(df2)
```
