
## Spark Joins Overview

### Internal Joins vs Actual Joins
- **Internal Joins**: Spark's physical join strategies (e.g., BroadcastHashJoin, SortMergeJoin, ShuffleHashJoin) chosen by the optimizer based on data size and join type.
- **Actual Joins**: The logical join types you specify in your code (e.g., inner, left, right, outer).

### Joining Two DataFrames in PySpark

python
df1.join(df2, on="key_column", how="inner")


### Types of Joins in PySpark

- **Inner Join**: `how="inner"`
- **Left Outer Join**: `how="left"`
- **Right Outer Join**: `how="right"`
- **Full Outer Join**: `how="outer"`
- **Left Semi Join**: `how="left_semi"`
- **Left Anti Join**: `how="left_anti"`

### Broadcast Join

- Use when one DataFrame is small enough to fit in memory.
- Spark sends the smaller DataFrame to all worker nodes to avoid shuffling.

python
from pyspark.sql.functions import broadcast
df1.join(broadcast(df2), on="key_column", how="inner")
