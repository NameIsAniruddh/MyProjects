
## Spark Joins Overview

### Internal Joins vs Actual Joins
- **Internal Joins**: Spark's physical join strategies (e.g., BroadcastHashJoin, SortMergeJoin, ShuffleHashJoin) chosen by the optimizer based on data size and join type.
- **Actual Joins**: The logical join types you specify in your code (e.g., inner, left, right, outer).

### Joining Two DataFrames in PySpark

python
df1.join(df2, (df1.col1 == df2.col2) & (df1.col3 == df2.col4), how="inner")


### Types of Joins in PySpark

- **Inner Join**: `how="inner"`
  python
  df1.join(df2, (df1.col1 == df2.col2) & (df1.col3 == df2.col4), how="inner")
  
  
- **Left Outer Join**: `how="left"`
  python
  df1.join(df2, (df1.col1 == df2.col2) & (df1.col3 == df2.col4), how="left")
  
  
- **Right Outer Join**: `how="right"`
  python
  df1.join(df2, (df1.col1 == df2.col2) & (df1.col3 == df2.col4), how="right")
  
  
- **Full Outer Join**: `how="outer"`
  python
  df1.join(df2, (df1.col1 == df2.col2) & (df1.col3 == df2.col4), how="outer")
  
  
- **Left Semi Join**: `how="left_semi"`
  python
  df1.join(df2, (df1.col1 == df2.col2) & (df1.col3 == df2.col4), how="left_semi")
  
  
- **Left Anti Join**: `how="left_anti"`
  python
  df1.join(df2, (df1.col1 == df2.col2) & (df1.col3 == df2.col4), how="left_anti")


### 4. Difference between `dropDuplicates`, `distinct`, and `delete` in PySpark (with example)

- **dropDuplicates**: Removes duplicate rows based on specified columns. Returns a new DataFrame; does not modify the original data in storage.

python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
data = [("A", 1), ("B", 2), ("A", 1), ("C", 3)]
df = spark.createDataFrame(data, ["name", "value"])

# Remove duplicate rows based on specific columns
df_no_duplicates = df.dropDuplicates(["name", "value"])
df_no_duplicates.show()


- **distinct**: Removes duplicate rows across all columns. Returns a new DataFrame; does not modify the original data in storage.

python
# Remove duplicate rows across all columns
df_distinct = df.distinct()
df_distinct.show()


- **delete**: Removes rows from a table based on a condition. Used in SQL or Delta tables; modifies the data in storage.

python
# PySpark DataFrame does not support `delete()` directly.
# To delete rows from a Delta table, use SQL or DeltaTable API.

from delta.tables import DeltaTable

# Example: Delete rows where value == 1 from a Delta table
delta_table = DeltaTable.forName(spark, "my_table")
delta_table.delete("value = 1")


**Summary**:
- `dropDuplicates`: Removes duplicates in a DataFrame (in-memory), based on specified columns, does not affect source table.
- `distinct`: Removes duplicates in a DataFrame (in-memory), across all columns, does not affect source table.
- `delete`: Removes rows from a table (persistent storage), based on a condition.

```python
data = [("A", 1, "X"), ("B", 2, "Y"), ("A", 1, "X"), ("C", 3, "Z"), ("A", 1, "W")]
df = spark.createDataFrame(data, ["name", "value", "category"])

# Remove duplicate rows based on specific columns
df_no_duplicates = df.dropDuplicates(["name", "value"])
display(df_no_duplicates)
```

```python
data = [("A", 1, "X"), ("B", 2, "Y"), ("A", 1, "X"), ("C", 3, "Z"), ("A", 1, "W")]
df = spark.createDataFrame(data, ["name", "value", "category"])

# Remove duplicate rows based on specific columns
df_no_distinct = df.distinct()
display(df_no_distinct)
```

```python
from pyspark.sql.functions import explode, col

# Sample input data
data = [
    {
        "datasets": [
            {
                "customerId": "C001",
                "orderId": "O1001",
                "orderDate": "2026-05-01",
                "orderDetails": {"item": "Laptop", "qty": 1},
                "shipmentDetails": {"carrier": "FedEx", "tracking": "TRK123"}
            },
            {
                "customerId": "C002",
                "orderId": "O1002",
                "orderDate": "2026-05-02",
                "orderDetails": {"item": "Phone", "qty": 2},
                "shipmentDetails": {"carrier": "UPS", "tracking": "TRK456"}
            }
        ]
    },
    {
        "datasets": [
            {
                "customerId": "C003",
                "orderId": "O1003",
                "orderDate": "2026-05-03",
                "orderDetails": {"item": "Tablet", "qty": 3},
                "shipmentDetails": {"carrier": "DHL", "tracking": "TRK789"}
            }
        ]
    }
]

ordersDf = spark.createDataFrame(data)

# Explode datasets and extract fields
parseOrdersDf = ordersDf.withColumn("orders", explode(col("datasets"))) \
    .withColumn("customerId", col("orders").getItem("customerId")) \
    .withColumn("orderId", col("orders").getItem("orderId")) \
    .withColumn("orderDate", col("orders").getItem("orderDate")) \
    .withColumn("orderDetails", col("orders").getItem("orderDetails")) \
    .withColumn("shipmentDetails", col("orders").getItem("shipmentDetails"))

display(parseOrdersDf)
```

```python
### 5. Difference between `sortBy`, `orderBy`, and `distributeBy` in PySpark

- **sortBy**: Used in RDDs to sort data based on a function. Returns a new RDD.
  python
  rdd = sc.parallelize([("A", 2), ("B", 1), ("C", 3)])
  rdd_sorted = rdd.sortBy(lambda x: x[1])
  print(rdd_sorted.collect())
  

- **orderBy**: Used in DataFrames to sort rows based on one or more columns. Returns a new DataFrame.
  python
  df.orderBy("column_name", ascending=True).show()
  

- **distributeBy**: Used to repartition DataFrame rows based on specified columns (for parallelism, not sorting).
  python
  df_distributed = df.distributeBy("column_name")
  

**Summary**:
- `sortBy`: Sorts RDDs using a function.
- `orderBy`: Sorts DataFrames by columns.
- `distributeBy`: Repartitions DataFrames for parallel processing, not sorting.
```
