Real‑time Example (Very Simple)
- DataFrame Example:
When you load a CSV of employee records, Spark treats each row as generic columns without enforcing a strict type.
Like reading an Excel sheet — flexible but not type‑safe.
- Dataset Example:
When you create a case class Employee(id: Int, name: String) and load the same data, Spark enforces the structure at compile time.
Like having a strongly typed Java class — safer and catches errors early.


```python
from pyspark.sql.functions import col

df = spark.read.format("csv").load("/Volumes/workspace/default/data_analysis/empdata.csv")
df_final = df.toDF("first_name", "last_name", "age", "doj", "salary")
df_final = df_final.withColumn("last_name", col("last_name").cast("int"))
display(df_final)
```
