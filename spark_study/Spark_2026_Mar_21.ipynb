
### Basic Idea about Partition

- **Partition**
  - Horizontal division, grouping, splitting, or separation of data.
  - Each partition contains a subset of the overall data.

- **Why Partition?**
  - To divide and distribute data for parallel processing.
  - Improves throughput and enables concurrent access/processing.
  - Essential for scalability and performance in distributed systems.

- **How to Control Partitioning**
  - Partitions are created naturally based on data source and cluster configuration.
  - You can change the number of partitions using:
    - `repartition`: Increases or decreases partitions, shuffles data.
    - `coalesce`: Reduces partitions efficiently, minimizes data movement.
  - Adjusting partitions controls the degree of parallelism.
