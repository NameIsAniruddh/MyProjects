```python
### Map-comprehensions and filter-comprehensions
nums = [1, 2, 3, 4, 5, 6]

# map using comprehension
mapped_result = [x * 10 for x in nums]

# filter using comprehension
filtered_result = [x for x in nums if x % 2 == 0]

# map + filter together
map_filter_result = [x * 10 for x in nums if x % 2 == 0]


print("Original List")
print(nums)
print()

print("Map Comprehension")
print(mapped_result)
print()

print("Filter Comprehension")
print(filtered_result)
print()

print("Map + Filter Comprehension")
print(map_filter_result)
```

```python
### List, Tuple, Set and Dictionary comprehensions
# Comprehensions in Python
nums = [1, 2, 3, 4, 5, 5, 2]

# List Comprehension
list_result = [x * 2 for x in nums]

# Tuple Comprehension
tuple_result = tuple(x * 2 for x in nums)

# Set Comprehension
set_result = {x * 2 for x in nums}

# Dictionary Comprehension
dict_result = {x: x * 2 for x in nums}

print("List Comprehension")
print(list_result)
print()
print("Tuple Comprehension")
print(tuple_result)
print()
print("Set Comprehension")
print(set_result)
print()
print("Dictionary Comprehension")
print(dict_result)
```

```python
def fib(n):
 first,second=0,1
 for i in range(n):
    yield first
    third=first+second
    first,second=second,third

gen = fib(10)
for value in gen:
    print(value)
```

```python
class fib:
    def __init__(self,n):
        self.n=n
        self.first,self.second=0,1
        self.i=0
    def __iter__(self):
        return self
    def __next__(self):
        if (self.i+1)>self.n:
            raise StopIteration
        self.i+=1
        value=self.first
        self.first,self.second=self.second,self.first+self.second
        return value   
fib_iter=fib(5)
for num in fib_iter:
    print(num)         
```

```python
class Account:
    def __init__(self, name, number, amount):
        self.name = name
        self.number = number
        self.balance = amount

    def __repr__(self):
        return f'[name={self.name}, number={self.number}, balance={self.balance}]'

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def transfer(self, amount, other):
        self.withdraw(amount)
        other.deposit(amount)
account1 = Account('John', 12345, 1000)
account2 = Account('Mary', 54321, 2000)
account1.transfer(500, account2)
print(account1, account2)
```

```python
# using static method and class method 
class Employee:
    company = "ABC Technologies"
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    # instance method
    def show_employee(self):
        print("Name    :", self.name)
        print("Salary  :", self.salary)
        print("Company :", Employee.company)

    # class method
    @classmethod
    def change_company(cls, new_company):
        cls.company = new_company

    # static method
    @staticmethod
    def is_valid_salary(salary):
        return salary > 0

emp1 = Employee("John", 50000)
emp2 = Employee("Mary", 60000)

print("Before Company Change")
print("---------------------")
emp1.show_employee()
print()
emp2.show_employee()


print()
print("Using Class Method")
print("------------------")

Employee.change_company("XYZ Solutions")
emp1.show_employee()
print()
emp2.show_employee()


print()
print("Using Static Method")
print("-------------------")
print(Employee.is_valid_salary(5000))
print(Employee.is_valid_salary(-100))
```

```python
# Inheritance and Duck-typing.
# Parent Class
class Animal:
    def sound(self):
        print("Animal makes sound")

# Child Classes
class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Duck Typing Class
class Robot:
    def sound(self):
        print("Robot speaks digitally")

# common function
def make_sound(obj):
    obj.sound()

dog = Dog()
cat = Cat()
robot = Robot()

make_sound(dog)
make_sound(cat)
make_sound(robot)
```

```python
import csv

class Product:
    def __init__(self, name, stock, price):
        self.name = name
        self.stock = int(stock)
        self.price = float(price)

    def __repr__(self):
        return f'Product(name={self.name}, stock={self.stock}, price={self.price})'

class ProductManager:
    def __init__(self, filename):
        self.filename = filename

    def load_products(self):
        products = []
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.DictReader(file)
                for row in reader:
                    products.append(Product(row['name'], row['stock'], row['price']))
        except FileNotFoundError:
            pass
        return products

    def save_products(self, products):
        with open(self.filename, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=['name', 'stock', 'price'])
            writer.writeheader()
            for p in products:
                writer.writerow({'name': p.name, 'stock': p.stock, 'price': p.price})

    def add_product(self, name, stock, price):
        products = self.load_products()
        products.append(Product(name, stock, price))
        self.save_products(products)

    def list_products(self):
        products = self.load_products()
        for p in products:
            print(p)

    def update_product(self, name, stock=None, price=None):
        products = self.load_products()
        for p in products:
            if p.name == name:
                if stock is not None:
                    p.stock = int(stock)
                if price is not None:
                    p.price = float(price)
        self.save_products(products)

    def delete_product(self, name):
        products = self.load_products()
        products = [p for p in products if p.name != name]
        self.save_products(products)

# Usage Example:
pm = ProductManager('products.csv')
pm.add_product('Apple', 100, 0.5)
pm.add_product('Banana', 50, 0.3)
pm.list_products()
pm.update_product('Apple', stock=120)
pm.delete_product('Banana')
pm.list_products()
```
