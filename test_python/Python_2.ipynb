```python
def find_sum_of_n(*nums):
    print(type(nums))
    result=0
    for num in nums:
        result+=num
    return result
print(find_sum_of_n(10))
print(find_sum_of_n(10,20,30))
```

```python
def lam_net(sal,hra_per=0,da_per=0):
    find_per_val= lambda base,per:(per/100)*base
    hra=find_per_val(sal,hra_per)
    da=find_per_val(sal,da_per)
    net_sal=sal+hra+da
    return net_sal
print(lam_net(300000,20,10))
```

```python
def fact(n):
    if n<=1:
     return 1
    else:
        return n*fact(n-1)
print(fact(5))
```
